```python
import pandas as pd

# 예: 'my_data.csv'라는 파일을 업로드했다면
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/J.유료서비스_월평균_이용금액_21-23.csv', encoding = 'cp949')

# 데이터 확인
df.head(28)
```

#지역별 데이터 삭제( 데이터 :유료서비스_월평균_이용금액 )

'유료서비스_월평균_이용금액' 데이터에서 실제 분석에 사용할 필요가 없다고 판단한, 지역별 요약 행(인덱스 2~18)을 삭제함


```python
# 예: 인덱스 5번부터 9번까지의 행 삭제
df = df.drop(df.index[2:19])
df.head()
```

```python
df.head(10)
```

#결측치 제거

`dropna()`를 사용하여 결측치가 있는 행이 제거되도록 처리함. 그 결과 제거된 행이 없어 결측치가 없는 것을 확인함


```python
df = df.dropna()
df.head()
```
