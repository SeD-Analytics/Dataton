### 점유율 계산


```python
import pandas as pd
import matplotlib.pyplot as plt

# 파일 불러오기
df = pd.read_csv(file_paths[2], header=[0, 1])

# 전체/소계 행 선택
row = df[(df[('구분별(1)', '구분별(1)')] == '전체') & (df[('구분별(2)', '구분별(2)')] == '소계')]

# 대상 OTT 및 연도
ott_list = ['넷플릭스 (%)', '웨이브 (%)', '티빙 (%)', '왓챠 (%)']
years = ['2021', '2022', '2023']

# 점유율 계산
data = {'Year': years}
for ott in ott_list:
    data[ott.replace(' (%)', '')] = []

for year in years:
    total = sum([row[(year, ott)].values[0] for ott in ott_list])
    for ott in ott_list:
        value = row[(year, ott)].values[0]
        data[ott.replace(' (%)', '')].append(round(value / total * 100, 2))  # 점유율 %

df_share = pd.DataFrame(data)

# ✅ 시각화: 점유율
plt.figure(figsize=(10, 6))
for col in df_share.columns[1:]:
    plt.plot(df_share['Year'], df_share[col], marker='o', label=col)

plt.title('OTT 플랫폼별 점유율 변화 (2021~2023)', fontsize=14)
plt.xlabel('연도')
plt.ylabel('점유율 (%)')
plt.ylim(0, 100)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
