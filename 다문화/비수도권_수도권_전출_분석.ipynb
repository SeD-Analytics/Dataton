```python
import pandas as pd   #라이브러리 pandas 호출
import numpy as np    #라이브러리 numpy 호출

from google.colab import files

# 파일 업로드
uploaded = files.upload()

# 업로드된 파일명을 자동으로 가져오기
filename = list(uploaded.keys())[0]

# 파일 불러오기
# 인코딩 명시
df = pd.read_csv(filename, encoding='cp949')
print(f"✅ 데이터 '{filename}' 로드 완료!")
df.head()
```

```python

capital_regions = ['서울특별시', '경기도', '인천광역시']
print(df['전입행정_시도_LABEL'].unique())

# 수도권으로 전입한 데이터 중 비수도권에서 온 것
to_capital_from_non = df[
    (df['전입행정_시도_LABEL'].isin(capital_regions)) &
    (~df['전출행정_시도_LABEL'].isin(capital_regions))
]

# 전출 지역별 이사 건수(행 수) 계산 후 내림차순 정렬
outflow_by_region = to_capital_from_non.groupby('전출행정_시도_LABEL').size().sort_values(ascending=False)

# 결과 출력
print(outflow_by_region)
```
